```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

goals = requests.get('https://www.premierleague.com/stats/top/players/goals')

goals_content = goals.content
soup_goal = BeautifulSoup(goals_content, 'html.parser')
```

```python
#Top 10 players
selected_group = soup_goal.select(".stats-table__name")
player_name = []

for name in selected_group:
    name_value = name.get_text()
    player_name.append(name_value)
    
spacing_fixed = []
final_fix = []
    
for players in player_name:
    spacing_fixed.append(players.replace('\n\n      ', ''))
    
for players in spacing_fixed:
    final_fix.append(players.replace('\n', ''))

#Top 10 goals
selected_goal = soup_goal.select(".stats-table__main-stat")
player_goal = []

for goal in selected_goal:
    goal_value = goal.get_text()
    player_goal.append(goal_value)

top_dict = {"Player": final_fix, "Goal":player_goal}

df = pd.DataFrame.from_dict(top_dict)
df
```
